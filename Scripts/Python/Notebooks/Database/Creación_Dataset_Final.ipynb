```python
url = "https://raw.githubusercontent.com/Sxmuu/TG-Samuel-P/main/Databases/Meteo/salidas%20NASA/nasa_power_daily.csv"

clima = pd.read_csv(url)

# === tu lista de estaciones ===
coords_alt_list = [
    ("Usaquen", 4.71035, -74.03042, 2570),
    ("Suba", 4.76125, -74.09346, 2571),
    ("Las Ferias", 4.69070, -74.08249, 2552),
    ("Centro de Alto Rendimiento", 4.65847, -74.08396, 2552),
    ("Fontibon", 4.67825, -74.14382, 2551),
    ("Puente Aranda", 4.63177, -74.11749, 2590),
    ("Kennedy", 4.62505, -74.16133, 2580),
    ("Tunal", 4.57623, -74.13096, 2589),
    ("San Cristobal", 4.57256, -74.08381, 2688),
    ("Jazmin", 4.60850, -74.11494, 2559),
    ("Usme", 4.53206, -74.11714, 2593),
    ("Ciudad Bolivar", 4.57781, -74.16628, 2661),
]

# === a DataFrame de estaciones ===
est = pd.DataFrame(coords_alt_list, columns=["Estacion", "lat", "lon", "Altitud"])

# === asegurar tipos numéricos en ambos DFs (por si venían como texto) ===
for df in (clima, est):
    df["lat"] = pd.to_numeric(df["lat"], errors="coerce")
    df["lon"] = pd.to_numeric(df["lon"], errors="coerce")

# === (sanidad) revisar duplicados por (lat, lon) en la tabla de estaciones ===
dup_count = est.duplicated(["lat", "lon"]).sum()
if dup_count:
    print(f"⚠️ Estaciones duplicadas por (lat, lon): {dup_count}")

# === merge exacto por lat/lon: añade Estacion y Altitud al DF climático ===
clima = clima.merge(est[["lat", "lon", "Estacion", "Altitud"]],
                    on=["lat", "lon"], how="left")

# === chequeos rápidos ===
print("Porcentaje sin estación asignada:", clima["Estacion"].isna().mean())
print(clima.head())
```

```python
import pandas as pd

# ======================
# 1) Partimos de df_latlon y clima ya cargados en memoria
# ======================

# --- A. Parseo de fechas (normalizadas a medianoche) ---
df_latlon["Date"] = pd.to_datetime(df_latlon["Date"], format="%Y-%m-%d", errors="coerce").dt.normalize()
clima["Date"]     = pd.to_datetime(clima["date"],     format="%d/%m/%Y", errors="coerce").dt.normalize()

# --- B. Aseguramos tipos numéricos en coordenadas ---
for df in (df_latlon, clima):
    df["lat"] = pd.to_numeric(df["lat"], errors="coerce")
    df["lon"] = pd.to_numeric(df["lon"], errors="coerce")

# ======================
# 2) Higiene del clima (deduplicar por si hay múltiples filas por misma llave)
# ======================
vars_clima = ["Rad","Precip","Hum","Temp","WindSpeed"]  # columnas climáticas
key_cols   = ["Date","lat","lon"]

# Si hay duplicados por (Date,lat,lon), consolidamos promediando las variables climáticas.
# Conservamos Estacion/Altitud si son consistentes; si no, tomamos la primera aparición.
if clima.duplicated(key_cols).any():
    clima_ag = (
        clima.groupby(key_cols, dropna=False, as_index=False)
             .agg({**{c: "mean" for c in vars_clima},
                   "Estacion": "first",
                   "Altitud": "first"})
    )
else:
    clima_ag = clima[key_cols + vars_clima + ["Estacion","Altitud"]].copy()

# ======================
# 3) Merge LEFT (contaminación ← clima)
# ======================
# Antes del merge, quita Estacion/Altitud del clima para no duplicar
clima_ag_nometas = clima_ag.drop(columns=["Estacion","Altitud"], errors="ignore")

df_final = df_latlon.merge(
    clima_ag_nometas, on=["Date","lat","lon"], how="left", validate="m:1"
)


# ======================
# 4) Chequeos rápidos
# ======================
print("Rango fechas contaminación:", df_latlon["Date"].min(), "→", df_latlon["Date"].max())
print("Rango fechas clima        :", clima_ag["Date"].min(), "→", clima_ag["Date"].max())
print("Nulos por variable climática (proporción):")
print(df_final[vars_clima].isna().mean())
print(df_final.head())
```

```python
# =========================================
# 1) Imports
# =========================================
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, explained_variance_score

# =========================================
# 2) Configuración
# =========================================
TARGET = "PM25"
# Features: clima + estáticas + calendáricas (sin PM10 contemporáneo)
BASE_FEATURES = ["lon", "lat", "Altitud", "Temp", "Rad", "Hum", "WindSpeed", "Precip", "month", "doy"]
SECTOR_COL = "Sectores-CLUSTERIN"
USE_SECTOR_DUMMIES = True

# Corte temporal para OOT (out-of-time)
TRAIN_END = pd.Timestamp("2023-12-31")   # <= entreno hasta aquí, evalúo en 2024

# =========================================
# 3) Funciones auxiliares
# =========================================
def parse_dates_inplace(df: pd.DataFrame, col_in: str = "Date"):
    if col_in not in df.columns and "date" in df.columns:
        df["Date"] = pd.to_datetime(df["date"], errors="coerce")
    elif col_in in df.columns:
        df["Date"] = pd.to_datetime(df[col_in], errors="coerce")
    else:
        raise ValueError("No encuentro columna de fecha (Date o date)")
    df["Date"] = df["Date"].dt.normalize()
    return df

def add_calendar_features(df: pd.DataFrame):
    df["month"] = df["Date"].dt.month.astype("int16")
    df["doy"]   = df["Date"].dt.dayofyear.astype("int16")
    return df

def ensure_numeric(df: pd.DataFrame, cols):
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df

def add_sector_dummies(df: pd.DataFrame):
    if (SECTOR_COL in df.columns) and USE_SECTOR_DUMMIES:
        dummies = pd.get_dummies(df[SECTOR_COL].astype("category"), prefix="SEC", dummy_na=False)
        df = pd.concat([df, dummies], axis=1)
    return df

def metricas(y_true, y_pred):
    mse  = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae  = mean_absolute_error(y_true, y_pred)
    r2   = r2_score(y_true, y_pred)
    evs  = explained_variance_score(y_true, y_pred)
    eps  = 1e-8
    mape = np.mean(np.abs((y_true - y_pred) / (np.abs(y_true) + eps))) * 100
    return {"MAE":mae, "MSE":mse, "RMSE":rmse, "MAPE":mape, "R2":r2, "EVS":evs}

def build_feature_list(df: pd.DataFrame):
    feats = BASE_FEATURES.copy()
    if USE_SECTOR_DUMMIES and SECTOR_COL in df.columns:
        feats += [c for c in df.columns if c.startswith("SEC_")]
    return feats

# =========================================
# 4) Preparar df_final (solo MÉTRICAS)
# =========================================
# Suponemos df_final ya unido (contaminación + clima)
df_final = parse_dates_inplace(df_final, "Date")
df_final = ensure_numeric(df_final, ["lat","lon","Altitud","Temp","Rad","Hum","WindSpeed","Precip", TARGET])
df_final = add_calendar_features(df_final)
df_final = add_sector_dummies(df_final)

# Filas válidas
df_ok = df_final.dropna(subset=[TARGET,"Temp","Rad","Hum","WindSpeed","Precip","lat","lon","Altitud"])

# Split temporal OOT
train_mask = df_ok["Date"] <= TRAIN_END
test_mask  = (df_ok["Date"] > TRAIN_END) & (df_ok["Date"] <= pd.Timestamp("2024-12-31"))

df_tr = df_ok.loc[train_mask].copy()
df_te = df_ok.loc[test_mask].copy()

FEATURES = build_feature_list(df_tr)

print(f"Entrenamiento hasta {TRAIN_END.date()}: {len(df_tr)} filas")
print(f"Prueba OOT en 2024: {len(df_te)} filas")
print(f"Features ({len(FEATURES)}): {FEATURES}")

# =========================================
# 5) Entrenar modelo y evaluar
# =========================================
model = xgb.XGBRegressor(
    objective="reg:squarederror",
    n_estimators=300,
    learning_rate=0.08,
    max_depth=5,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42
)
model.fit(df_tr[FEATURES], df_tr[TARGET])

# ---- Eval 1: Out-of-time (2024) ----
if len(df_te) > 0:
    pred_te = model.predict(df_te[FEATURES])
    m_oot = metricas(df_te[TARGET], pred_te)
    print("\n[Métricas OOT (2024)]")
    for k,v in m_oot.items():
        print(f"{k}: {v:.4f}")
else:
    print("\nNo hay filas para OOT 2024 con el corte actual.")

# ---- Eval 2: Holdout temporal interno (20% final del train) ----
df_tr_sorted = df_tr.sort_values("Date")
cut = int(len(df_tr_sorted) * 0.8)
df_tr_sub = df_tr_sorted.iloc[:cut]
df_va_sub = df_tr_sorted.iloc[cut:]

m_hold = xgb.XGBRegressor(
    objective="reg:squarederror",
    n_estimators=300, learning_rate=0.08, max_depth=5,
    subsample=0.9, colsample_bytree=0.9, random_state=42
)
m_hold.fit(df_tr_sub[FEATURES], df_tr_sub[TARGET])
pred_va = m_hold.predict(df_va_sub[FEATURES])
m_hold_metrics = metricas(df_va_sub[TARGET], pred_va)

print("\n[Métricas Holdout temporal (último 20% del tren)]")
for k,v in m_hold_metrics.items():
    print(f"{k}: {v:.4f}")
```
